```python
def data_separtaion(input,rows):
    attackdata=list()
    normaldata=list()
    avg=st.mean
    sd=st.stdev
    attackclass=[]
    normalclass=[]
    
    for i in range(rows):
        
        if input.iat[i,10]==0:
           # print(input.iat[0,10])
            normaldata.append(input.iloc[i])
        else:
            attackdata.append(input.iloc[i])

    attackrows=len(attackdata)
    normalrows=len(normaldata)
    print(len(attackdata),len(normaldata))
    for i in zip(*attackdata):
     
      attackclass.append((avg(i),sd(i)))
    
    for i in zip(*normaldata):
        normalclass.append((avg(i),sd(i)))

    print(len(attackclass),len(normalclass))

    normalclass=normalclass[:-1]
    attackclass=attackclass[:-1]
    
    return attackclass,normalclass,attackrows,normalrows


 
def class_probability(attackrows,normalrows,totalrows):
     print(attackrows,normalrows,totalrows)
     attackprobability=attackrows/totalrows
     normalprobability=normalrows/totalrows
     return attackprobability,normalprobability

def predict(attackclass,normalclass,row,attackprobability,normalprobability):
    for i in attackclass:
       prior=gauss_distrib (i[0],i[1],row[attackclass.index(i)])
       attackprobability*=prior
    #print(attackprobability)
    for i in normalclass:
        likelihood=gauss_distrib (i[0],i[1],row[normalclass.index(i)])
        #print(prior)
        normalprobability*=likelihood
    #print(normalprobability)
    dic=dict()
    dic[0]=normalprobability
    dic[1]=attackprobability
    predictlabel=""
    probability=0
    for label,value in dic.items():
        if probability<value:
            predictlabel=label
            probability=value
    return predictlabel


def gauss_distrib(mean,stddev,x):
   pi=math.pi
   cal=-((x-mean)**2)/(2*(stddev)**2)
   #Sprint(cal)
   power=math.exp(cal) 
   prob=1/((stddev*math.sqrt(2*pi)))*power
   return prob

def trainset():

    df=X_train
    rows,columns=df.shape
    attackclass,normalclass,attackrows,normalrows=data_separtaion(df,rows)
    #print(attackclass)
    #print(normalclass)
    attackprobability,normalprobability=class_probability(attackrows,normalrows,rows)
    #print(attackprobability,normalprobability)
    return attackclass,normalclass,attackprobability,normalprobability

def testset(attackclass,normalclass,attackprobability,normalprobability):
    test=X_test
    rowslength,columnslength=test.shape
    #predict(attackclass,normalclass,row,attackprobability,normalprobability)
    actuallabel=list()
    predictlabel=list()
    for i in range(rowslength):
        predictlabel.append(predict(attackclass,normalclass,test.iloc[i,:],attackprobability,normalprobability))

    actuallabel=test['attack_class'].tolist()
    print(len(actuallabel))
    print(len(predictlabel))
    return actuallabel,predictlabel
def accuracy(actuallabel,predictlabel):
    accuracy=0
    TP=0
    TN=0
    FP=0
    FN=0
    Confusion_Matrix=[2,2]
    for i in range(len(actuallabel)):
        if actuallabel[i]==predictlabel[i]:
            accuracy+=1
        if actuallabel[i]==predictlabel[i]==0:
            TP+=1
        elif actuallabel[i]==0 and predictlabel[i]!=0:
            FP+=1
        elif actuallabel[i]==1 and predictlabel[i]!=1:
            FN+=1
        elif actuallabel[i]==predictlabel[i]==1:
            TN+=1

    Accuracy=accuracy/(len(actuallabel))
    Precision=(TP/(TP+FP))
    Recall=(TP/(TP+FN))
    F1_Score=2*((Precision*Recall)/(Precision+Recall))
    Confusion_Matrix=[[TP,FP],[FN,TN]]
    print(Confusion_Matrix)
    print(Accuracy,TP+TN+FP+FN,Precision,Recall,F1_Score)
    return Accuracy,Precision,Recall,F1_Score,Confusion_Matrix
```

```python
attackclass,normalclass,attackprobability,normalprobability=trainset()
actuallabel,predictlabel=testset(attackclass,normalclass,attackprobability,normalprobability)
Accuracy,Precision,Recall,F1_Score,Confusion_Matrix=accuracy(actuallabel,predictlabel)
print("Accuracy:",Accuracy)
print("Precision:",Precision)
print("Recall:",Recall)
print("F1_Score:",F1_Score,)
print("Confusion_Matrix:",Confusion_Matrix)
```
